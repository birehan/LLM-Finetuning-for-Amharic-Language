## Amharic Vocabulary Dataset

```python
dic_path = "../data/am_ET/am_ET.txt"
```

```python
# Open the text file in read mode
vocabulary = []

with open(dic_path, 'r') as file:
    # Read lines from the file
    lines = file.readlines()
    # Append each line to the vocabulary list
    for line in lines:
        vocabulary.append(line.strip())  # strip() removes leading and trailing whitespace

# Now vocabulary contains each line of the file as an element

```
